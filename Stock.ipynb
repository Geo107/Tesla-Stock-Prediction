```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
ts = pd.read_csv('TSLA.csv',index_col=0)
ts
```

```python
ts['MA10'] = ts['Close'].rolling(10).mean()
ts['MA50'] = ts['Close'].rolling(50).mean()
ts = ts.dropna()
```

```python
ts['Shares'] = [1 if ts.loc[ei, 'MA10']>ts.loc[ei, 'MA50'] else 0 for ei in ts.index]
```

```python
ts.head
```

```python
ts['Close1'] = ts['Close'].shift(-1)
ts
```

```python
ts['Profit'] = [ts.loc[ei, 'Close1'] - ts.loc[ei, 'Close'] if ts.loc[ei, 'Shares']==1 else 0 for ei in ts.index]

ts
```

```python
ts['Profit'].plot()
plt.axhline(y=0, color='red')
```

```python
ts['wealth'] = ts['Profit'].cumsum()
ts.tail()
```

```python
ts['wealth'].plot()
plt.title('Total money you win is {}'.format(ts.loc[ts.index[-2], 'wealth']))
```
